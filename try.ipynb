```python
import cv2
import numpy as np

def preprocess_image(input_path, output_path='preprocessed.jpg'):
    # Load image
    img = cv2.imread(input_path)

    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Light sharpening (optional)
    kernel = np.array([[0, -1, 0], [-1, 5,-1], [0, -1, 0]])
    sharpened = cv2.filter2D(gray, -1, kernel)

    # Resize if image is very large (keep aspect ratio)
    max_dim = 1024
    h, w = sharpened.shape
    if max(h, w) > max_dim:
        scale = max_dim / max(h, w)
        sharpened = cv2.resize(sharpened, (int(w*scale), int(h*scale)))

    # Save preprocessed image
    cv2.imwrite(output_path, sharpened)
    return output_path

# Preprocess the image
preprocessed_path = preprocess_image('MD_13.jpg')
```

```python
import ollama

# Path to your preprocessed image
# image_path = 'MD_13.jpg'
image_path = 'preprocessed.jpg'

# Read the image as bytes
with open(image_path, 'rb') as f:
    image_bytes = f.read()

stop_tokens = [
    "<|im_start|>",
    "<|im_end|>"
]

response = ollama.chat(
    model='minicpm-v:8b',
    messages=[
        {
            'role': 'user',
            'content': 'Please perform OCR on this image',
            'images': [image_bytes]
        }
    ],
    options={
        "stop": stop_tokens
    }
)

print(response['message']['content'])
```
